```python
import numpy as np
import pandas as pd
```

```python
dict1 = {
    'name':['harry','rohan','skillf','shubh'],
    'marks':[92,34,24,17],
    'city':['rampur','kolkata','bareilly','antartica'],
}
```

```python
# it converts the dict to like an excel sheet for faster indexing
df = pd.DataFrame(dict1)
```

```python
df
# the key of the dict becomes column names and the each column is indexed
```

```python
# we are exporting this dataframe to csv file
df.to_csv('friends.csv')
```

```python
# we dont want the index to be present in the csv file
df.to_csv('friends.csv',index=False)
```
